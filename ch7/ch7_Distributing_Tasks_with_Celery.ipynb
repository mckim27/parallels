# 7. Distributing Tasks with Celery

### - Understanding Celery

Celery 는 분산시스템을 손쉽게 구축할 수 있도록 도와주는 프레임워크. 
Celery 프레임워크는 로컬 또는 네트워크 컴퓨터 간에 메시지를 교환하는 work units (tasks) 의 분산개념으로 동작한다.
task 가 celery 의 key concept

<p><br/><br/></p>

### - Why use Celery? 

다음의 여러 장점들로 인해 celery 를 사용한다.

• 로컬 또는 네트워크 간에 worker 들이 투명한 방식으로 배포된다.

• 설정을 통해 간단한 방법으로 worker 들의 병행성을 제어한다.  

• 동기, 비동기, 주기적이거나 예약된 작업을 지원한다. 

• 에러가 발생했을 경우 재실행한다.

### - Working with tasks  

아래 코드는 위의 그림에서 본다면 worker 측에서 수행.

```python
from celery import Celery

app = Celery('helloWorker',
             broker='amqp://',
             backend='amqp')
@app.task
def hello_world():
    return "Hello I'm a celery task"
```

- delay(arg, kwarg=value) : apply_async method 의 약식.


- apply_async((arg,), {'kwarg': value}) : 비동기식으로 broker 에게 데이터를 전달.

        위의 그림으로 따지면 client 에서 delay, apply_async 같은 method를 호출.
        ex) object.apply_async((data1, data2), queue='testqueue', countdown=1, expires=10)

- apply((arg,), {'kwarg': value}) : 동기식으로 broker에게 데이터를 전달.


- <a href='http://celery.readthedocs.org/en/latest/reference/celery.app.task.html'>method 자세히 보기</a>

### - Understanding workers

celery 는 worker를 알아서 컨트롤함.

다음과 같은 방법들로 worker를 컨트롤 한다.

- Concurrency mode: 이것은 worker 가 수행되는 모드, 예를 들어, Process,Thread, Eventlet 및 Gevent


- Remote control: 이 메커니즘을 사용하여, 우리는 특정 worker 나 여러 worker에게 직접 메시지를 보낼 수 있다. 이런 worker 을 통해 특정 queue 에 접근하게할 수도 있다.


-  Revoking tasks: 이 메커니즘을 통해 우리는 하나 혹은 하나 이상의 worker 에게 특정 task를 무시하게 할 수 있다.

### - Dispatching a simple task (ex)

- 아래 파일의 이름을 tasks.py 로 작성.

```python
from math import sqrt
from celery import Celery

app = Celery('tasks', broker='redis://192.168.0.6:6379/0', backend='redis://192.168.0.6:6379/0')
#app.config.CELERY_RESULT_BACKEND = 'redis://192.168.0.6:6379/0'

@app.task
def sqrt_task(value):
  return sqrt(value)
```

* celery worker -A tasks -l info 명령으로 실행 ( foreground 로 실행.)
    * ' -A ' 옵션은 '--app' 와 같음, 위에서 작성한 파일 이름과 -A 옵션 다음에 나오는 이름이 같아야 함. 
    * ' -l ' 옵션은 '--loglevel' 과 같음

```python
import time
from tasks import *

logging.basicConfig(level=logging.DEBUG)


def manage_sqrt_task(value):
  result = app.send_task('tasks.sqrt_task', args=(value,),
                   queue='sqrt_queue', routing_key='sqrt_queue')
  logging.info(result.get())
  
  
if __name__ == '__main__':
  manage_sqrt_task(4)
 
```

### - Using Celery to obtain a Fibonacci series term (ex)

```python
# -*- coding: utf-8 -*-

from math import sqrt
from celery import Celery
import logging

app = Celery('tasks', broker='redis://127.0.0.1:6379/0', backend='redis://127.0.0.1:6379/0')
#app.config.CELERY_RESULT_BACKEND = 'redis://192.168.0.6:6379/0'

@app.task
def sqrt_task(value):
  return sqrt(value)

@app.task
def fibo_task(value):
  a, b = 0,1
  for item in range(value):
    a, b = b, a + b
  message = "The Fibonacci calculated with task id %s was %d" % (fibo_task.request.id, a)
  return (value, message)
```

```python
import logging
import time
from tasks import *

logging.basicConfig(level=logging.DEBUG)


input_list = [4, 3, 8, 6, 10]


def manage_fibo_task(value_list):
  async_result_dict = {x: app.send_task('tasks.fibo_task', args=(x,), queue='fibo_queue',
                            routing_key='fibo_queue') for x in value_list}
  time.sleep(2)
  for key, value in async_result_dict.items():
    if value.ready():
      logging.info("Value [%d] -> %s" % (key,  value.get()[1]))
    else:
      logging.info("Task [%s] is not ready" % value.task_id)


if __name__ == '__main__':
  manage_fibo_task(input_list)
```

### - Using Celery to make a distributed Web crawler (ex) 

```python
# -*- coding: utf-8 -*-

from math import sqrt
from celery import Celery
import logging
import re
import requests

html_link_regex = re.compile(
 '<a\s(?:.*?\s)*?href=[\'"](.*?)[\'"].*?>')

app = Celery('tasks', broker='redis://127.0.0.1:6379/0', backend='redis://127.0.0.1:6379/0')
#app.config.CELERY_RESULT_BACKEND = 'redis://192.168.0.6:6379/0'

@app.task
def sqrt_task(value):
  return sqrt(value)

@app.task
def fibo_task(value):
  a, b = 0,1
  for item in range(value):
    a, b = b, a + b
  message = "The Fibonacci calculated with task id %s was %d" % (fibo_task.request.id, a)
  return (value, message)

@app.task
def crawl_task(url):
  request_data = requests.get(url)
  links = html_link_regex.findall(request_data.text)
  message = "The task %s found the following links %s.."
  return message
```

<p><br/><br/></p>

```python
# -*- coding: utf-8 -*-

import logging
import time
from tasks import *

logging.basicConfig(level=logging.DEBUG)

url_list = ['http://www.google.com',
            'http://br.bing.com',
            'http://duckduckgo.com',
            'http://github.com',
            'http://br.search.yahoo.com']


def manage_crawl_task(url_list):
  async_result_dict = {url: app.send_task('tasks.crawl_task', args=(url,), queue='webcrawler_queue',
                              routing_key='webcrawler_queue') for url in url_list}
  time.sleep(4)
  for key, value in async_result_dict.items():
    if value.ready():
      logging.info("%s -> %s" % (key, value.get()))
    else:
      logging.info("The task [%s] is not ready" % value.task_id)


if __name__ == '__main__':
  manage_crawl_task(url_list)
```

### - Background 로 실행하기

```python
celery multi start w1 –A tasks –Q sqrt_queue,fibo_queue,webcrawler_queue -l info -c 2
```

* w1 은 node 이름. 


* -A 는 app 지정. 


* -Q 는 queue 지정.


* -l 은 log 옵션.


* -c 는 concurrency 옵션 프로세스의 갯수 를 지정 가능.


* 여러 개의 worker로도 실행 가능.
   * celery multi start <b>2</b> –A tasks –Q sqrt_queue,fibo_queue,webcrawler_queue -l info -c 2
   * celery multi start <b>w1 w2</b> –A tasks –Q sqrt_queue,fibo_queue,webcrawler_queue -l info -c 2
   * 위의 예시들의 경우 총 tasks 들의 수는 worker 갯수 x process의 갯수 = 4
